```python
def hist(df, col):
    plt.figure(figsize=(12,7))
    sns.histplot(df[col].dropna(), kde=True, bins='auto', color='orange')
    plt.ylabel('count')
    plt.xlabel(f'{col}')
    plt.title(f'{col}의 히스토그램')
    plt.tight_layout()
    return plt.show()

lst = ['dti', 'fico_range_low', 'fico_range_high', 'delinq_2yrs', 'acc_now_delinq', 'delinq_amnt', 'inq_last_6mths',
       'mths_since_last_delinq', 'mths_since_last_record', 'mths_since_last_major_derog', 'collections_12_mths_ex_med',
       'chargeoff_within_12_mths', 'pub_rec_bankruptcies', 'tax_liens']

for col in lst:
    hist(df_choice,col)
```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# 수치형 컬럼만 선택
num_cols = df_choice.select_dtypes(include="number").columns.tolist()

chunk_size = 5  # 한 번에 5개 컬럼씩
n_cols = 5      # 한 행에 5개
n_rows = 1

for i in range(0, len(num_cols), chunk_size):
    cols_chunk = num_cols[i:i+chunk_size]
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 3))
    if n_cols == 1:
        axes = [axes]
    
    for j, col in enumerate(cols_chunk):
        sns.boxplot(y=df_choice[col], ax=axes[j])
        axes[j].set_title(col, fontsize=10)
        axes[j].set_xlabel("")
    
    # 남는 축 숨기기
    for j in range(len(cols_chunk), n_cols):
        axes[j].axis("off")
    
    plt.tight_layout()
    plt.show()
```
