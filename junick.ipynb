```python
# decile 생성
df["dti_decile"] = pd.qcut(df["dti"], 10, duplicates="drop")

# 구간별 bad rate
dti_bad = df.groupby("dti_decile")["mask"].mean()

# 시각화
plt.figure(figsize=(6,4))
dti_bad.plot(marker="o")
plt.title("Bad Rate by DTI Decile")
plt.ylabel("Bad Rate")
plt.xlabel("DTI Decile")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

```python
vars_to_check = [
    "int_rate",
    "dti",
    "fico_mid",
    "revol_util",
    "installment_to_income"
]

for col in vars_to_check:
    # 1) 필요한 컬럼만 가져오기
    temp = df[[col, "mask"]].copy()

    # 2) 숫자로 변환 (문자열/이상값 있으면 NaN으로)
    temp[col] = pd.to_numeric(temp[col], errors="coerce")

    # 3) inf / -inf 제거 (installment_to_income 같은 파생변수에서 자주 발생)
    temp = temp.replace([np.inf, -np.inf], np.nan)

    # 4) NaN 제거
    temp = temp.dropna(subset=[col, "mask"])

    # 5) 유니크 값이 너무 적으면 decile(10분위) 불가능 → 스킵
    if temp[col].nunique() < 10:
        print(f"[SKIP] {col}: unique values = {temp[col].nunique()} (10분위 불가)")
        continue

    # 6) decile 생성
    temp["decile"] = pd.qcut(temp[col], 10, duplicates="drop")

    # 7) 구간별 bad rate 계산
    bad_rate = temp.groupby("decile")["mask"].mean()

    # 8) 시각화
    plt.figure(figsize=(6,4))
    bad_rate.plot(marker="o")
    plt.title(f"Bad Rate by {col} Decile")
    plt.ylabel("Bad Rate")
    plt.xlabel(f"{col} Decile")
    plt.xticks(rotation=45)
    plt.grid(alpha=0.3)
    plt.show()
```

```python
# 1) E/F/G만 선택
df_efg = df[df["grade"].isin(["E", "F", "G"])].copy()

# 2) 주요 수치형 컬럼 숫자형 변환(문자/이상값 방어)
num_candidates = [
    "fico_range_low", "fico_range_high",
    "installment", "annual_inc",
    "loan_amnt",
    "revol_bal", "revol_util"
]
for c in num_candidates:
    if c in df_efg.columns:
        df_efg[c] = pd.to_numeric(df_efg[c], errors="coerce")

# revol_util이 "55.3%" 같은 문자열이면 % 제거 후 숫자화 (이미 숫자면 그대로)
if "revol_util" in df_efg.columns and df_efg["revol_util"].dtype == "object":
    df_efg["revol_util"] = (
        df_efg["revol_util"].astype(str).str.replace("%", "", regex=False)
    )
    df_efg["revol_util"] = pd.to_numeric(df_efg["revol_util"], errors="coerce")

# 3) 파생변수 생성
df_efg["fico_mid"] = (df_efg["fico_range_low"] + df_efg["fico_range_high"]) / 2

df_efg["monthly_inc"] = df_efg["annual_inc"] / 12
df_efg["installment_to_income"] = df_efg["installment"] / df_efg["monthly_inc"]
df_efg["loan_to_income"] = df_efg["loan_amnt"] / df_efg["annual_inc"]

df_efg["revol_bal_to_income"] = df_efg["revol_bal"] / df_efg["annual_inc"]

# revol_pressure: revol_util * revol_bal
# revol_util이 0~100(%) 형태라면 scale 영향이 크니, 보통은 /100 해서 비율로 쓰는 걸 추천
df_efg["revol_util_ratio"] = df_efg["revol_util"] / 100
df_efg["revol_pressure"] = df_efg["revol_util_ratio"] * df_efg["revol_bal"]

# 4) inf/-inf 제거
df_efg = df_efg.replace([np.inf, -np.inf], np.nan)

```

```python
# 비교할 수치형 컬럼 목록
num_cols = [
    "fico_mid", "annual_inc", "installment_to_income", "loan_to_income",
    "revol_bal_to_income", "revol_util", "revol_pressure",
    "loan_amnt", "installment", "revol_bal"
]
num_cols = [c for c in num_cols if c in df_efg.columns]  # 존재하는 컬럼만

# 0/1 그룹 분리
paid = df_efg[df_efg["mask"] == 0]
bad  = df_efg[df_efg["mask"] == 1]

# 그룹별 평균/중앙값
mean_paid = paid[num_cols].mean(numeric_only=True)
mean_bad  = bad[num_cols].mean(numeric_only=True)

median_paid = paid[num_cols].median(numeric_only=True)
median_bad  = bad[num_cols].median(numeric_only=True)

# 전체 표준편차(표준화 차이 계산용)
std_all = df_efg[num_cols].std(numeric_only=True)

# 요약표 만들기
summary_num = pd.DataFrame({
    "mean_paid(0)": mean_paid,
    "mean_bad(1)": mean_bad,
    "diff(0-1)": mean_paid - mean_bad,
    "median_paid(0)": median_paid,
    "median_bad(1)": median_bad
})

# 표준화된 차이 크기(절댓값): 클수록 두 그룹을 더 잘 가름
summary_num["std_diff_abs"] = (summary_num["diff(0-1)"] / std_all).abs()

# 차이가 큰 순으로 보기
summary_num_sorted = summary_num.sort_values("std_diff_abs", ascending=False)

display(summary_num_sorted)
```
